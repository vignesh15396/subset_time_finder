```python
import pandas as pd
```

```python
df=pd.read_excel('inp.xlsx',header=None)
df
```

```python
df[['user','duration','start','end','unknown']]=df[0].str.split('|').to_list()
df.drop(labels=[0],axis=1,inplace=True)
df
```

```python
df1=df.copy()
df1=df1.sort_values(by='start').reset_index(drop=True)
df1
```

```python
# startnum and endnum columns are used to find the overlap of time
df1['startnum']=df1.start.str.replace(':','')
df1['startnum']=df1.startnum.str.replace('-','')
df1['startnum']=df1.startnum.str.replace(' ','')
df1['endnum']=df1.end.str.replace(':','')
df1['endnum']=df1.endnum.str.replace('-','')
df1['endnum']=df1.endnum.str.replace(' ','')
df1['startnum']=pd.to_numeric(df1.startnum)
df1['endnum']=pd.to_numeric(df1.endnum)
# df1['start']=pd.to_datetime(df1['start'])
# df1['end']=pd.to_datetime(df1['end'])
df1
```

```python
users=df1.user.unique()
#ud dictionary will have 4 lists of all start time, endtime, startnum, endnum for each user
ud={}
for i in users:
    
    ud[i]=[]
    ud[i].append(list(df1[df.user==i].start))
    ud[i].append(list(df1[df.user==i].end))
    ud[i].append(list(df1[df.user==i].startnum))
    ud[i].append(list(df1[df.user==i].endnum))    
    
ud
```

```python
#ufin dictionary will have selected time intervals after eliminating the overlap time and subset time 
ufin={}
#below for will iterate through unique users
for i in users:
    ufin[i]=[]
    #below for will iterate through each user's records
    for j in range(len(ud[i][0])): 
        #first time interval of the user taken by using if,since the ufin dictionary is empty
        if len(ufin[i])==0:
            ufin[i].append([ud[i][0][j],ud[i][1][j],ud[i][2][j],ud[i][3][j]])
        #fi-flag for time interval which is a subset of already selected time interval of a user
        #fm-flag for time interval overlaping with already selected time interval of a user
        fi=0
        fm=0
        #below for will iterate through each user's selected time intervals
        for k in range(len(ufin[i])):    
            #if start and end of current time lies inside the already selected time interval, fi flag raised, the current time will be ignored
            if ud[i][2][j] in range(ufin[i][k][2],ufin[i][k][3]) and ud[i][3][j] in range(ufin[i][k][2],ufin[i][k][3]):   
                fi=1
            #if start of current time is inside the already selected time interval but end not inside, then change selected end = current end and fm flag raised
            #ex:if already selected 1 to 5, current 2 to 7 then selected will be updated as 1 to 7
            elif ud[i][2][j] in range(ufin[i][k][2],ufin[i][k][3]):
                ufin[i][k][1]=ud[i][1][j]
                fm=1
        #if the user's time interval neither fall inside nor overlap with already selected time interval then update the ufin with current time interval
        if fi==0 and fm==0:
            ufin[i].append([ud[i][0][j],ud[i][1][j],ud[i][2][j],ud[i][3][j]])        

#time intervals after eliminating subset time and overlap time     
ufin
```

```python
#after getting the selected list, converting it to dataframe
from numpy import array
temp_list=[]
for i in users:
    for j in range(len(ufin[i])):
        temp_list.append([i]+ufin[i][j])
temp_arr=array(temp_list).T
temp_arr
```

```python
data = {'user':temp_arr[0],'start':temp_arr[1],'end':temp_arr[2]}
data
```

```python
dffin=pd.DataFrame(data)
dffin['start']=pd.to_datetime(dffin['start'])
dffin['end']=pd.to_datetime(dffin['end'])
#finding the duration
dffin['total']=dffin['end']-dffin['start']
dffin
```

```python
#finding the total duration of each user
df_total=dffin.groupby('user')['total'].sum().copy()
for i in range(len(df_total)):
    print('user={} Total hours spend = {} hrs'.format(df_total.index[i],df_total[i]))
```
